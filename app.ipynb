```python
import os
```

```python
import nltk
import textract
```

```python
import ipywidgets as widgets
from IPython.display import HTML
```

```python
nltk.download('punkt')
```

```python
from nltk.tokenize import sent_tokenize
```

```python
def get_text_from_pdf(text_file):
    return textract.process(text_file).decode("utf-8")
```

```python
def get_sentences(text):
    return len(sent_tokenize(text))
```

```python
uploader = widgets.FileUpload(
    accept=".pdf", 
    multiple=False
)
```

```python
text_input = widgets.Textarea(
    value='',
    placeholder='Type something',
    description='Text:',
    disabled=False
)
```

```python
out = widgets.Output()
```

```python
display(uploader, text_input, out)
```

```python
def run_refresh_upload(*ignore):
    input_file = "test.pdf"
    for f in uploader.value.keys():
        file_name, file_extension = os.path.splitext(f)
        if file_extension == ".pdf":
            with open(input_file, "wb") as fp:
                fp.write(uploader.value[f]["content"])
            with out:
                display(get_sentences(text=get_text_from_pdf(text_file=input_file)))
            os.remove(input_file)
        else: 
            print('Can not convert: ', f)
```

```python
def run_refresh_text(*ignore):
    input_file = "test.pdf"
    with out:
        display(get_sentences(text=text_input.value))
```

```python
text_input.observe(run_refresh_text, 'value')
```

```python
uploader.observe(run_refresh_upload, 'value')
```
